```python
import pandas as pd

xgb = pd.read_csv('xgb_predictions_tune.csv')
lgb = pd.read_csv('lgb_predictions (2).csv')
cat = pd.read_csv('cat_predictions_tune.csv')

# ID 기준으로 merge (concat이 아니라 merge를 쓰는 이유: 행 순서가 파일마다 다를 수 있어서)

merged = xgb.merge(lgb, on='ID').merge(cat, on='ID')

print(merged.shape)  # 90,067행이 나와야 정상
merged.head()
```

```python
merged['final_pred'] = (merged['xgb_pred'] + merged['lgb_pred'] + merged['cat_pred']) / 3
```

```python
sample = pd.read_csv('sample_submission.csv')  # 대회/부트캠프에서 받은 제출 양식

#sample_submission의 ID 순서 그대로, final_pred 값만 채워넣기

submission = sample.merge(merged[['ID', 'final_pred']], on='ID', how='left')
submission[sample.columns[1]] = submission['final_pred']  # 정답 컬럼명에 맞춰 채움
submission = submission[sample.columns]  # 원래 컬럼 순서/구성으로 정리

submission.to_csv('final_submission.csv', index=False)
```
